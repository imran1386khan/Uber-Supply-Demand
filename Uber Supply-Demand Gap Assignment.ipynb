**Import Essential Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import time
from datetime import datetime
import datetime as dt
```

**Import CSV**

```python
df = pd.read_csv("Uber Request Data.csv")
```

**Data Inspection**

```python
df.info()
df.describe()
```

**Data Type Conversion**

```python
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'])
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'])
```

**Create Derived Attribute**

```python
df['Trip time minutes'] = df.apply(lambda x: x['Drop timestamp'] - x['Request timestamp'], axis=1)
minutes = []
for x in df['Trip time minutes']:
    minutes.append(round(x.seconds / 60, 0))
df['Trip time minutes'] = minutes
```

```python
df['Week day num'] = df['Request timestamp'].apply(lambda x: datetime.isoweekday(x))
```

```python
df['Week day name'] = df['Request timestamp'].apply(lambda x: x.strftime("%A"))
```

```python
df['Month day num'] = df['Request timestamp'].apply(lambda x: x.day)
```

```python
df['Hour of day'] = df['Request timestamp'].apply(lambda x: x.hour)
```

```python
def time_slots(hr):
    if hr<5:
        a = "Early_Morning"
    elif hr>=5 and hr<10:
        a = "Morning"
    elif hr>=10 and hr<18:
        a = "Day_Time"
    elif hr>=17 and hr<20:
        a = "Evening"
    else:
        a = "Late_Night"
    return a
```

```python
##applying time Slots function for estimating the slot during the day cab was booked. 
df['Time slot'] = df['Hour of day'].apply(lambda x: time_slots(x))
```

**WeekDay Visualizations**

```python
plt.figure(figsize=(6,4))
sns.set(style="darkgrid")
ax = sns.countplot(x="Week day name", hue="Status", data=df)
plt.title("WeekDay Status")
```
